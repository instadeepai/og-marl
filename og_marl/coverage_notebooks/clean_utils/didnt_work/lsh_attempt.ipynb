```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# first, randomly generate hyperplanes

dim = 2
n_datapoints = 1000

# maxes = np.random.randint(5,20,size=dim)
# mins = np.zeros(dim)

random_hyperplane = np.random.uniform(-1,1,size=dim)

vectors = np.random.randint(-20,20,size=(dim,n_datapoints))

random_bias_unscaled = np.random.uniform(0,1)

# normalise all incoming data?

mins = [min(vectors[i,:]) for i in vectors.shape[0]]
maxes = [max(vectors[i,:]) for i in vectors.shape[0]]
scales = mins*maxes

random_bias = [x*(max(vectors[i,:])-min(vectors[i,:]))+min(vectors[i,:]) for i,x in enumerate(random_bias_unscaled)]
```

```python
hashval = np.tensordot(vectors,random_hyperplane,axes=[0,0])
```
